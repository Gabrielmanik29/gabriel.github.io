### Download dan Muat Data

```python
import kagglehub
import pandas as pd
import os

path = kagglehub.dataset_download("michaelmatta0/movies-ultimate-metrics-features-and-metadata")

print("Path to dataset files:", path)

for file in os.listdir(path):
    if file.endswith(".csv") and file.startswith("Top Movies"):
        dataset_path = os.path.join(path, file)
        break


if dataset_path:
    data = pd.read_csv(dataset_path)
    print(data.head())
else:
    print(f"Error: Could not find the CSV file in {path}")
```
